````python
from tardis_data_downloader.data.data_manager import TardisDataManager, TardisApi
````

````python
from dotenv import load_dotenv, find_dotenv
import os
load_dotenv(find_dotenv())
if os.getenv("TARDIS_API_KEY") is None:
    from getpass import getpass
    tardis_api_key = getpass()
else:
    tardis_api_key = os.getenv("TARDIS_API_KEY")
````

````python
exchanges = TardisApi().get_exchanges()
exchanges
````

````python
exchange_ids = [exchange['id'] for exchange in exchanges]
exchange_ids
````

````python
manager = TardisDataManager(exchange="deribit", api_key=tardis_api_key)
manager
````

````python
manager.list_exchanges()
````

````python
manager.list_exchange_symbols()
````

````python
manager.batch_download_by_date_range(
    data_types=['trades'],
    symbols=['BTC-PERPETUAL'],
    start_date='2025-08-01',
    end_date='2025-08-03',
    sequential=False,
)
````
